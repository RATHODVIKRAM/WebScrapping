```python
import pandas as pd
import time
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome()  
url = "https://www.screener.in/screens/71064/all-stocks/?page="
company_lists = []
for i in range(1,196):
    url1 = url + str(i)
    driver.get(url1)
    time.sleep(1)
    html_content = driver.page_source
    #print(html_content)
    soup = BeautifulSoup(html_content,'html.parser')
    table = soup.find('table')
    rows = table.find_all('tr')[1:]
    for row in rows:
        columns = row.find_all('td')
        if columns:
            companydata = [col.text.strip() for col in columns]
            company_lists.append(companydata)
print(company_lists)
print(len(company_lists))
if company_lists:
    df = pd.DataFrame(company_lists, columns=["s.no","Company Name","current price","price to earning","market capitalization","dividend yield","net profit latest quater","YOY quarterly profit growth","sales latest quater","YOY quarterly sales growth","return on capital employed"])
    df.to_csv("b_company_names.csv", index=False)
    print("Data has been saved to company_names_a.csv.")
else:
    print("No data found.")
```
