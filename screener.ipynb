```python
import pandas as pd
import time
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome()  
url = "https://www.screener.in/screens/71064/all-stocks/?page="
company_names = []
for i in range(1,196):
    url1 = url + str(i)
    driver.get(url1)
    time.sleep(1)
    html_content = driver.page_source
    soup = BeautifulSoup(html_content,'html.parser')
    table = soup.find('table')
    
    rows = table.find_all('tr')[1:]
    for row in rows:
        columns = row.find_all('td')
        if columns:
            #print(f"Columns found: {[col.text.strip() for col in columns]}")
            company_names_col = columns[1].find('a')
            if company_names_col:
                company_name = company_names_col.text.strip()
                company_names.append(company_name)
            else:
                company_names.append(columns[1].text.strip())
print(len(company_names))
print(f"company names: ",company_names)
if company_names:
    df = pd.DataFrame(company_names, columns=["Company Name"])

    df.to_csv("a_company_names.csv", index=False)
    print("Data has been saved to company_names_a.csv.")
else:
    print("No data found.")
```
